```python
import sqlite3

def display_anki_flashcards(pqb_file):
    # Connect to the SQLite database
    conn = sqlite3.connect(pqb_file)
    cursor = conn.cursor()

    # Query to select flashcards
    query = """
    SELECT
        notes.id,
        notes.sfld AS question,
        notes.flds AS answer
    FROM
        notes
    """

    cursor.execute(query)
    flashcards = cursor.fetchall()

    # Display flashcards
    for flashcard in flashcards:
        print(f"ID: {flashcard[0]}")
        print(f"Question: {flashcard[1]}")
        print(f"Answer: {flashcard[2]}")
        print("-" * 20)

    # Close the connection
    conn.close()

# Example usage
pqb_file = 'Pleco Flash Backup 250101.pqb'
display_anki_flashcards(pqb_file)
```
